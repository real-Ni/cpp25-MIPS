```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(4):
    data = np.loadtxt(f"snapshot_0.{2*(i+1)}.dat")

    x = data[:, 0]
    y = data[:, 1]
    theta = data[:, 2]

    u = np.cos(theta)
    v = np.sin(theta)

    plt.figure(figsize=(5, 5))
    plt.quiver(x, y, u, v, angles="xy", scale_units="xy", scale=100)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"snapshot @ rho = 0.{2*i}")
    plt.show()
```
