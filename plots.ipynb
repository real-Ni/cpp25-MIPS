```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('fmax_vs_pe.dat', sep=' ')
plt.figure(figsize=(5, 4))
plt.plot(df['Pe'], df['fmax'], 'ro-')
plt.xlabel('PÃ©clet Number (Pe)')
plt.ylabel('Largest Cluster Fraction')
plt.title('MIPS Phase Transition')
plt.savefig('plot_mips.png')
```

```python
msd = np.loadtxt('msd_data.dat')
plt.figure(figsize=(5, 4))
plt.loglog(msd[:, 0], msd[:, 1], label='Simulation')
plt.loglog(msd[:, 0], msd[:, 0]**2 * 50, '--', label='~t^2 (Ballistic)')
plt.loglog(msd[:, 0], msd[:, 0] * 200, ':', label='~t (Diffusive)')
plt.legend()
plt.title('MSD Check')
plt.savefig('plot_msd.png')
```

```python
snap = np.loadtxt('snapshot.dat')
plt.figure(figsize=(5, 5))
plt.scatter(snap[:, 0], snap[:, 1], s=5)
plt.title('MIPS Snapshot (Pe=150)')
plt.savefig('plot_snapshot.png')
```
