```python
L = 100
omega = 0.01
z = 1 + omega + np.sqrt(omega**2 + 2*omega)

# 
p_z = 1 - z**2
p_dash_z = (1 - z) * p_z / (1 + z)
q_z = (1 - z) * (1 - z) * (1 - pow(z, L))
delta_z = 2 * (1 + z) * (z - pow(z, L))

Z = 4 * (delta_z + (L - 1) * q_z)
n = np.linspace(1, L-1, L-1)
```

```python
Ppp_n_list = []
Ppm_n_list = []
for i in n:
    Ppp_n = (p_z * (pow(z, i) + pow(z, L-i)) + q_z) / Z

    Ppm_n = (p_dash_z * (pow(z, i) - pow(z, L-i)) + q_z) / Z
    if i == 1:
        Ppm_n += delta_z / Z

    Ppp_n_list.append(Ppp_n)
    Ppm_n_list.append(Ppm_n)
Vpp = -np.log(np.array(Ppp_n_list))
Vpm = -np.log(np.array(Ppm_n_list))
```

```python
plt.figure(figsize=(8, 6))
plt.plot(n, Vpp, label='-ln(Ppp(n))')
plt.plot(n, Vpm, label='-ln(Ppm(n))', color='green')
plt.xlabel('Particle Separation n')
plt.ylabel('Effective Potential V(n)')
plt.title('Effective Interaction Potentials between Particles')
plt.xlim(0, 50)
plt.ylim(1, 8)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
traj_rat = np.loadtxt('trajectory_rat.dat') # run and tumble motion
traj_srw = np.loadtxt('trajectory_srw.dat') # symmetric random walk
time_rat = traj_rat[:, 0]
positions_rat = traj_rat[:, 1:]

time_srw = traj_srw[:, 0]
positions_srw = traj_srw[:, 1:]
plt.plot(positions_rat, time_rat, 'k.', ms=0.5)
plt.xlabel('Positions')
plt.ylabel('Time')
plt.title('Run and Tumble')
plt.tight_layout()
plt.show()

plt.plot(positions_srw, time_srw, 'k.', ms=0.5)
plt.xlabel('Positions')
plt.ylabel('Time')
plt.title('Symmetric Random Walk')
plt.tight_layout()
plt.show()
```
